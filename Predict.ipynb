```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('home_dataset.csv')

# View the first few rows of the dataset
display(df.head())

# Extract target information
sizes = df['HouseSize'].values
prices = df['HousePrice'].values
```

```python
# Data visualization
plt.scatter(sizes, prices, marker='s', color='blue' )
plt.title('House Size vs House Price')
plt.xlabel('Size of House')
plt.ylabel('Price of House')
plt.show()
```

```python
# Split the data into training and testing sets
sizes_train, sizes_test, prices_train, prices_test = train_test_split(sizes, prices, test_size=0.2, random_state=42)

# Reshape the data for sklearn
sizes_train = sizes_train.reshape(-1, 1)
sizes_test = sizes_test.reshape(-1, 1)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(sizes_train, prices_train)
```

```python
# Make predictions on the test set
predictions = model.predict(sizes_test)

# Visualize the results
plt.scatter(sizes_test, prices_test, color='blue', label="Actual Prices")
plt.plot(sizes_test, predictions, color='red', label='Predictions', linewidth=2)
plt.title('Property Price Prediction with Linear Regression')
plt.xlabel('Size of House')
plt.ylabel('Price of House')
plt.legend()
plt.show()


```
